```python
def get_dataset(path, label):
    return [[path+x, label] for x in os.listdir(path) if not x.startswith('.')]


# Data Augmentation
def process_data(root_train, root_val, batch_size):
    #to prevent overfitting
    train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=90,
                                   width_shift_range=100,
                                   height_shift_range=100,
                                   horizontal_flip=True,
                                   vertical_flip=False,
                                   zoom_range=0.3)


    validation_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(root_train,
                                                        target_size=(ig_size, ig_size),
                                                        batch_size=batch_size,class_mode='categorical',
                                                       shuffle=True,color_mode='rgb')

    validation_generator = validation_datagen.flow_from_directory(root_val,
                                                                  target_size=(ig_size, ig_size),
                                                                  batch_size=batch_size,
                                                                  class_mode='categorical',
                                                                 shuffle=False,color_mode='rgb')



    return train_generator, validation_generator 


#add top layer
def addTopModel(bottom_model, num_classes):
    top_model = bottom_model.output
    top_model = GlobalAveragePooling2D()(top_model)
    top_model = Dense(128, activation='relu', kernel_initializer='he_normal')(top_model)
    top_model = Dropout(0.3)(top_model)
    top_model = Dense(num_classes, activation='sigmoid')(top_model)
    return top_model

    
#confusion matrix 
# def plot_confusion_matrix(cm, names, title='Confusion matrix', 
#                             cmap=plt.cm.Blues):
#     plt.imshow(cm, interpolation='nearest', cmap=cmap)
#     plt.title(title)
#     plt.colorbar()
#     tick_marks = np.arange(len(names))
#     plt.xticks(tick_marks, names, rotation=45)
#     plt.yticks(tick_marks, names)
#     plt.tight_layout()
#     plt.ylabel('True label')
#     plt.xlabel('Predicted label')
    

def plot_confusion_matrix(model, test_gen):
    test_gen.reset()
    classes = ['COVID','NORMAL','PNEUMONIA']
    tick_marks = [0.5,1.5,2.5]

    
    Y_pred = model.predict(test_gen, test_gen.samples // test_gen.batch_size+1)
    y_pred = np.argmax(Y_pred, axis=1)
    print('Confusion Matrix')
    cm = confusion_matrix(test_gen.classes, y_pred)

    target_names =  ['COVID','NORMAL','PNEUMONIA']
    
    sns.heatmap(cm,cmap='Purples', annot=True)
    plt.xticks(tick_marks, classes)
    plt.yticks([0,1,2], classes)
    plt.title('Confusion Matrix')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
    acc = accuracy_score(test_gen.classes, y_pred)
    precision = precision_score(test_gen.classes, y_pred, average=None)
    recall = recall_score(test_gen.classes, y_pred, average=None)
    f1 = f1_score(test_gen.classes, y_pred, average=None)
    

    print("                 [COVID-19     NORMAL    PNEUMONIA]")
    print("Precision Score: {}".format(precision))
    print("Recall Score: {}".format(recall))
    print("F1 Score: {}".format(f1))
    print("Accuracy Score: {}".format(acc))

    return plt.show()
    
    
#model loss
def plot_learning_curves(r):
    plt.figure(figsize=(12,4))
    
    plt.subplot(1,2,1)
    plt.plot(r.history['loss'])
    plt.plot(r.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.grid(True)
    
    plt.subplot(1,2,2)
    plt.plot(r.history['accuracy'])
    plt.plot(r.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    
    return plt.show()



# y_predi = logreg.predict(X_test)
# y_testi = y_test
# lb = LabelBinarizer()
# lb.fit(y_predi)
# y_predi = lb.transform(y_pred)

# print('Log Loss: {}'.format(metrics.log_loss(y_testi, y_predi)))

# print('AUC: {}'.format(multiclass_roc_auc_score(y_test, y_pred)))

# skplt.metrics.plot_roc_curve(y_testi, y_predi)
# plt.legend(loc='lower left', bbox_to_anchor=(1.1, -0.1), ncol=1, fancybox=True)
# plt.show()

def plot_auc(y_test, y_pred, average="micro"):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    return roc_auc_score(y_test, y_pred, average=average)


#ROC AUC
def multiclass_roc_auc_score(test_gen, model, average="micro"):
    from sklearn import preprocessing
    test_gen.reset()
    classes = ['COVID','NORMAL','PNEUMONIA']
    lb = preprocessing.LabelBinarizer()
    tick_marks = np.arange(len(classes))

    
    Y_pred = model.predict(test_gen, test_gen.samples // test_gen.batch_size+1) 
    y_pred = np.argmax(Y_pred, axis=1)
    
    y_test = test_gen.classes

    
    # lb.fit(y_test)
    y_predi = model.predict(test_gen, test_gen.samples // test_gen.batch_size+1)
    y_testi = y_test
    lb = LabelBinarizer()
    lb.fit(y_pred)
    y_predi = lb.transform(y_pred)
    y_label = to_categorical(y_predi)
    
#     y_predi = pd.DataFrame(y_predi, columns = ['COVID-19', 'NORMAL', 'PNEUMONIA'])
#     y_test = pd.DataFrame(y_test, columns = ['COVID-19', 'NORMAL', 'PNEUMONIA'])
    
    print('Log Loss: {}'.format(metrics.log_loss(y_testi, y_predi)))

    print('AUC: {}'.format(plot_auc(y_test, y_pred)))

        

#     fpr_keras, tpr_keras, thresholds_keras = scikitplot.metrics.plot_roc(y_testi, y_predi)
    auc_keras = plot_auc(y_test, y_pred)

    scikitplot.metrics.plot_roc(y_testi, y_predi)
   
    plt.plot(color='darkorange', label='ROC curve (area = %0.2f)' % auc_keras)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.title('ROC curve')
    plt.xlim([0.0, 1.2])
    plt.ylim([0.0, 1.05])
    
    
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.legend(loc='lower left', bbox_to_anchor=(1.3, .6), ncol=1, fancybox=True)
    plt.grid(True) 

    return plt.show()
```

```python
def get_model(transferleaner, train_gen, test_gen):
    m = transferleaner
    pred = addTopModel(m, num_classes) #MobileNet = bottom_model
    model = Model(inputs = m.input, outputs = pred) #bottom layers, top layer


    #data augmentation
    train_generator, validation_generator = process_data(root_train, root_val, batch_size)


    model.summary()

    checkpoint = ModelCheckpoint(data_dir+'InceptionV3'+version+'.h5',
                                 monitor='val_loss',
                                 mode="min",
                                 save_best_only=True,
                                 verbose=1)

    earlystop = EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=10,
                              verbose=1,
                              restore_best_weights=True)

    learning_rate_reduction = ReduceLROnPlateau(monitor='val_loss',
                                                patience=10,
                                                verbose=1,
                                                factor=0.8,
                                                min_lr=0.0001,
                                                mode="auto",
                                                min_delta=0.0001,
                                                cooldown=5)

    callbacks = [checkpoint, earlystop, learning_rate_reduction]

    model.compile(loss=models_loss,
                  optimizer=models_opt,
                  metrics=['accuracy'])
    
    
    model.fit_generator(train_generator,
                                  steps_per_epoch=len_train_total//batch_size,
                                  epochs=epochs,
                                  callbacks=callbacks,
                                  validation_data=validation_generator,
                                  validation_steps=len_val_total//batch_size)

    history = model.fit_generator(train_generator,
                                  steps_per_epoch=len_train_total//batch_size,
                                  epochs=epochs,
                                  callbacks=callbacks,
                                  validation_data=validation_generator,
                                  validation_steps=len_val_total//batch_size)
  

    
    return model, history
```

```python
m = InceptionV3
pred = addTopModel(m, num_classes)
model = Model(inputs = m.input, outputs = pred) #bottom layers, top layer


#data augmentation
train_generator, validation_generator = process_data(root_train, root_val, batch_size)


model.summary()

checkpoint = ModelCheckpoint(data_dir+'InceptionV3'+version+'.h5',
                             monitor='val_loss',
                             mode="min",
                             save_best_only=True,
                             verbose=1)

earlystop = EarlyStopping(monitor='val_loss',
                          min_delta=0,
                          patience=10,
                          verbose=1,
                          restore_best_weights=True)

learning_rate_reduction = ReduceLROnPlateau(monitor='val_loss',
                                            patience=10,
                                            verbose=1,
                                            factor=0.8,
                                            min_lr=0.0001,
                                            mode="auto",
                                            min_delta=0.0001,
                                            cooldown=5)

callbacks = [checkpoint, earlystop, learning_rate_reduction]

model.compile(loss=models_loss,
              optimizer=models_opt,
              metrics=['accuracy'])


model.fit_generator(train_generator,
                              steps_per_epoch=len_train_total//batch_size,
                              epochs=epochs,
                              callbacks=callbacks,
                              validation_data=validation_generator,
                              validation_steps=len_val_total//batch_size)

history = model.fit_generator(train_generator,
                              steps_per_epoch=len_train_total//batch_size,
                              epochs=epochs,
                              callbacks=callbacks,
                              validation_data=validation_generator,
                              validation_steps=len_val_total//batch_size)
```

```python
## Confusion matrix
plot_confusion_matrix(model, validation_generator)
## Learning curve
plot_learning_curves(history)
## ROC AUC
multiclass_roc_auc_score(validation_generator, model)
```
